```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
US=pd.read_csv('likelihood-of-us-domestic-travel-due-to-covid-19-as-of-march-2020.csv')
US1 = US.rename(columns={"Unnamed: 1": "Percentage of interviewees"})
US1
US1=US1.drop([0])
US1
US1=US1.T
US1.columns = US1.iloc[0]
US1
US1.iloc[1,0:4].plot(kind='pie', autopct='%.2f',
                     title="Likelihood of U.S. domestic travel due to COVID-19 as of March 2020", 
                     fontsize=10,figsize=(5, 5),subplots=True)
```

```python
SG_Hstats=pd.read_csv("Time_Series_Hotel Statistics_Aug_20.csv")
SG_Hstats.iloc[13:0:-1].plot(x="Monthly", y="Standard Average Occupancy Rate",
                             title="Standard Average Occupancy Rate",color='blue', 
                             kind="bar",figsize=(7,5))
SG_Hstats.iloc[13:0:-1].plot(x="Monthly", y="Standard Average Room Rate",
                             title="Standard Average Room Rate",color='blue', 
                             kind="bar",figsize=(7,5))
```

```python
SG_receipts=pd.read_csv("tourism_receipt.csv")
SG_receipts=SG_receipts.rename(index={0:"Tourism Receipts",1: 'Accommodation',2:'Food & Beverage',3:'Shopping',4:'Sightseeing, Entertainment & Gaming',5:'Others'})
SG_receipts=SG_receipts.drop(["Variables"],axis='columns')
SG_receipts
SG_receipts = SG_receipts.replace(',','', regex=True)
SG_receipts = SG_receipts.replace('','_', regex=True)
SG_receipts=SG_receipts.astype(float)
SG_receipts
plt.style.use('ggplot')
SG_receipts.T.iloc[-10:,1:6].plot(kind="bar",figsize=(15,5), title="Singapore Tourism Receipt")
```

```python
Jap_opinion=pd.read_csv("opinion-on-domestic-travels-after-lifting-covid-19-restrictions-in-japan-2020.csv")
Jap_opinion.columns = [col.replace(' ','_') for col in Jap_opinion.columns]
Jap_opinion = Jap_opinion.rename(columns={"Unnamed:_1": "Percentage of interviewees"})
Jap_opinion=Jap_opinion.drop([0,1])
Jap_opinion=Jap_opinion.drop(["Unnamed:_2"],axis=1)
Jap_opinion=Jap_opinion.T
header=Jap_opinion.iloc[0]
Jap_opinion.columns = Jap_opinion.iloc[0]
Jap_opinion.iloc[1,1:6].plot(kind='pie', autopct='%.2f', fontsize=10,
                             title='Opinion_on_domestic_travels_after_lifting_COVID-19_restrictions_in_Japan_2020',
                             figsize=(6, 6))
```

```python
data_revenuechange=pd.read_csv("change_in_revenue_global.csv")
revenuechangedf=pd.DataFrame(data_revenuechange)
revenuechangedf
revenuechangedf.loc[:,"Revenue"].str.replace(',', '').astype(float)
revenuechangedf.Revenue=revenuechangedf.loc[:,"Revenue"].str.replace(',', '').astype(float) 
revenuechangedf.Revenue.plot(kind="bar",x="Year",rot=0,color=['blue','blue','grey'],title="Forecasted change in revenue from the travel and tourism industry due to the coronavirus in million USD")
plt.xticks(revenuechangedf.index,revenuechangedf["Year"].values)
```

```python
df6=pd.read_csv('total-visitor-international-arrivals-monthly.csv')
df6
df6[df6.month=='2000-01'] # 264
df6[df6.month=='2005-01'] # 324
df62=df6.iloc[264:324,:]
df62
fig,ax = plt.subplots()
ax.plot(df62.month,df62.total_international_visitor_arrivals)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.set_title('Total International Visitors in SG from 2001-2005')
plt.show()
```

```python
df6[df6.month=='2003-01'] # 300
df6[df6.month=='2003-12'] # 311
df63=df6.iloc[300:311,:]
df63
fig,ax = plt.subplots()
ax.plot(df63.month,df63.total_international_visitor_arrivals)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.set_title('Total International Visitors in SG in 2003')
plt.show()
```
